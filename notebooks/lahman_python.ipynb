```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# establish a database connection
engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/lahman_baseball")
```

```python
# use the connection to run a query using pandas!
batting = pd.read_sql("SELECT * FROM batting;", con=engine)
batting.head()
```

```python
people = pd.read_sql("SELECT * FROM people;", con=engine)
collegeplaying = pd.read_sql("SELECT * FROM collegeplaying;", con=engine)
schools = pd.read_sql("SELECT * FROM schools;", con=engine)
salaries = pd.read_sql("SELECT * FROM salaries;", con=engine)
fielding = pd.read_sql("SELECT * FROM fielding;", con=engine)
teams = pd.read_sql("SELECT * FROM teams;", con=engine)
```

```python
merged = people.merge(collegeplaying, left_on ='playerid', right_on = 'playerid',how='left', indicator = False)
merged = merged.merge(schools, left_on ='schoolid', right_on = 'schoolid',how='left', indicator = False)
merged = merged.merge(salaries, left_on ='playerid', right_on = 'playerid',how='left', indicator = False)
merged = merged.merge(fielding, left_on ='playerid', right_on = 'playerid',how='left', indicator = False)

merged
```

```python
#filter where id is Vanderbilt
vandy = merged[merged.schoolid == 'vandy']
#choosen columns
vandy = vandy[['playerid','namefirst','namelast', 'salary']].drop_duplicates()
#sum for each player
vandy= vandy.groupby(['namefirst','namelast'], as_index=False).sum()
#max earner
vandy = vandy.sort_values('salary', ascending=False)

vandy
```

# Question 2
a. Using the fielding table, group players into three groups based on their position:<br> label players with position OF as "Outfield",<br>
those with position "SS", "1B", "2B", and "3B" as "Infield",<br>
and those with position "P" or "C" as "Battery". <br>
b. Determine the number of putouts made by each of these three groups in 2016.<br>

```python
#cleaned for question
position = merged[['namefirst','namelast', 'pos', 'po','yearid']]
#find where conditions are met
position['pos'] = position['pos'].replace(['OF'],'Outfield')
position['pos'] = position['pos'].replace(['SS','1B','2B','3B'],'Infield')
position['pos'] = position['pos'].replace(['P','C'],'Battery')   
position
```

```python
#filter year
year = float('2016')
position_2016 = position[position['yearid'] == year].drop_duplicates()
#groupby and remove year
position_po_total = position_2016.groupby('pos').sum().drop(['yearid'],axis =1)

position_po_total
```

```python
# df cleaned for question
winning = teams[['teamid','wswin','yearid','w']]
#filter for: did not win, range
not_winning = winning[winning['wswin'] == 'N']
not_winning = not_winning[not_winning['yearid'].between(1970,2017)]
#sort values 
not_winning = not_winning.sort_values('w', ascending=False)

not_winning
```

```python
# df from last part - winning
#filter for: did win, range
not_winning_s = winning[winning['wswin'] == 'Y']
not_winning_s = not_winning_s[not_winning_s['yearid'].between(1970,2017)]
#sort values 
not_winning_s = not_winning_s.sort_values('w')

not_winning_s

```

Updated dataset does not have problem year.

```python
#most wins won vs when they did not
#filter for: range, max wins per year
in_range = winning[winning['yearid'].between(1970,2017)]
won_max = in_range.groupby(['yearid'])['w'].max().reset_index()
won_most = pd.merge(in_range, won_max, how='inner', on =['w','yearid'])
# of teams with most wins per year count who won and who lost
max_won = won_most[won_most['wswin'] == 'Y']
max_lost = won_most[won_most['wswin'] == 'N']
count_max_won = len(max_won)
count_max_lost = len(max_lost)
#find percentage
percent_max_won_ws = round(count_max_won / (count_max_won + count_max_lost) * 100,2)
    
answer = pd.DataFrame({'count_max_won':[count_max_won],'count_max_lost':[count_max_lost],'percent_max_won_ws':[percent_max_won_ws]})
answer
```
